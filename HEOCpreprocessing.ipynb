```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
```

```python
data = pd.read_csv("heloc_dataset_v1.csv")
```

```python
data.head(100)
```

```python
data.columns
```

```python
y,x = data.values[:,0],data.values[:,1:]
```

```python
y
```

```python
for n,i in enumerate(y):
    if i == "Bad":
        y[n] = 0
    else:
        y[n] = 1
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
x_norm = min_max_scaler.fit_transform(x)
```

```python
scale_data = np.hstack((y.reshape(len(y),1),x_norm))
```

```python
scale_data
```

```python
df = pd.DataFrame(scale_data)
```

```python
df = df.sample(frac=1).reset_index(drop=True)
```

```python
train = df[2000:]
test = df[:2000]
```

```python
df
```

```python
train.to_csv("Htrain.csv",header = False, index = False)
```

```python
test.to_csv("Htest.csv",header = False, index = False)
```
